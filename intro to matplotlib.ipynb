# MatplotLib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplot

```python
# 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows= 2, ncols = 2, figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# 2
fig,ax = plt.subplots(nrows =2,
                     ncols=2,
                     figsize=(10,5))
# plot to each fifferent index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

# Matplotlib OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# matplotlib oo method with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45,100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Choleserol");
# add a legend
ax.legend(*scatter.legend_elements(),title="Target");
#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
plt.style.use('seaborn-pastel')
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#Add data ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#customize
ax0.set(title="Heart Disease and Choleserol Levels",
       ylabel="Choleserol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add data ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
#customize
ax1.set(title="Heart Disease and Maximum Heart Rate",
       xlabel="Age",
       ylabel="Maximum Heart Rate")
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
#add a title to the figure
fig.suptitle("Heart Disease analysis", fontsize=20, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid')
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer") #changes the color scheme
# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Choleserol");
# add a legend
ax.legend(*scatter.legend_elements(),title="Target");
#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
plt.style.use('seaborn-whitegrid')
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#Add data ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="Accent")
ax0.set_xlim([50,80])
#customize
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add data ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="Accent")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#customize
ax1.set(title="Heart Disease and Maximum Heart Rate",
       xlabel="Age",
       ylabel="Maximum Heart Rate")
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
#add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```
